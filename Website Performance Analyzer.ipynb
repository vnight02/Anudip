```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("website_response_time.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
average_response_time = data['response_time_ms'].mean()
average_response_time
```

```python
mean = data['response_time_ms'].mean()
std_dev = data['response_time_ms'].std()
threshold = mean + 2 * std_dev
outliers = data[data['response_time_ms'] > threshold]
outliers
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['response_time_ms'], bins=10)
plt.xlabel("Response Time (ms)")
plt.ylabel("Frequency")
plt.title("Website Response Time Distribution")
plt.show()
```
